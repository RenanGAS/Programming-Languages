```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)
```

# GCC/G++ & Clang/Clang++
## N-body

```python
# nbody.c

x_gcc = [100, 1000, 100000, 1000000, 10000000, 100000000]
y_gcc = [0.003, 0.004, 0.030, 0.268, 2.587, 25.641]

x_clang = [100, 1000, 100000, 1000000, 10000000, 100000000]
y_clang = [0.003, 0.004, 0.032, 0.246, 2.376, 23.509]

for i in range(0, len(x_gcc)):
    plt.plot(x_gcc[i:], y_gcc[i:], 'ro-')
    
for i in range(0, len(x_clang)):
    plt.plot(x_clang[i:], y_clang[i:], 'bo-')
    
plt.ylabel("Tempo em segundos")
plt.xlabel("NÂº elementos")

gcc = mpatches.Patch(color='red', label='gcc')
clang = mpatches.Patch(color='blue', label='clang')

plt.legend(handles=[gcc, clang])

plt.figure(figsize=(10, 8))

plt.show()
```

## Binary Trees

```python
# binary_trees.cpp

x_cpp = [2, 4, 8, 16, 20]
y_cpp = [0.001, 0.001, 0.006, 3.717, 74.393]

x_clang = [2, 4, 8, 16, 20]
y_clang = [0.001, 0.001, 0.005, 2.796, 57.822]

for i in range(0, len(x_cpp)):
    plt.plot(x_cpp[i:], y_cpp[i:], 'ro-')
    
for i in range(0, len(x_clang)):
    plt.plot(x_clang[i:], y_clang[i:], 'bo-')
    
plt.ylabel("Tempo em segundos")
plt.xlabel("Profundidade MÃ¡xima")

cpp = mpatches.Patch(color='red', label='cpp')
clang = mpatches.Patch(color='blue', label='clang')

plt.legend(handles=[cpp, clang])

plt.figure(figsize=(10, 8))

plt.show()
```

O programa N-body Ã© composto de vÃ¡rias operaÃ§Ãµes matemÃ¡ticas, sendo do tipo CPU-bound. JÃ¡ o Binary Trees envolve a construÃ§Ã£o de vÃ¡rias Ã¡rvores binÃ¡rias, fazendo sucessivos acessos de memÃ³ria, sendo do tipo IO bound. Ao executarmos seus cÃ³digos na linguagem C, percebemos que hÃ¡ uma diferenÃ§a, mesmo que pouca, entre utilizarmos o compilador GCC e o compilador Clang. Isso acontece pelo Clang utilizar LLVM (Low Level Virtual Machine) para realizar e otimizar o processo de compilaÃ§Ã£o (anÃ¡lise lÃ©xica, sintÃ¡tica e semÃ¢ntica). Ã dito que o Clang trata o "front-end" e o LLVM, o "back-end", enquanto o GCC lida com ambos.

# CPython & PyPy
## N-body

```python
# nbody.py

x_python = [100, 1000, 100000, 1000000, 10000000]
y_python = [0.023, 0.027, 0.804, 7.691, 67.946]

x_pypy = [100, 1000, 100000, 1000000, 10000000]
y_pypy = [0.052, 0.065, 0.125, 0.665, 6.153]

for i in range(0, len(x_python)):
    plt.plot(x_python[i:], y_python[i:], 'ro-')
    
for i in range(0, len(y_pypy)):
    plt.plot(x_pypy[i:], y_pypy[i:], 'bo-')
    
plt.ylabel("Tempo em segundos")
plt.xlabel("NÂº elementos")

python = mpatches.Patch(color='red', label='python')
pypy = mpatches.Patch(color='blue', label='pypy')

plt.legend(handles=[python, pypy])

plt.figure(figsize=(10, 8))

plt.show()
```

Utilizando a linguagem Python no programa N-body, temos entÃ£o dois interpretadores: o CPython e o PyPy. Nesse caso, percebemos uma grande diferenÃ§a de tempos de execuÃ§Ã£o entre o compilador padrÃ£o (CPython) e o PyPy. Isso acontece pelo PyPy utilizar compilaÃ§Ã£o Just-In-Time, otimizando a execuÃ§Ã£o do cÃ³digo atravÃ©s da compilaÃ§Ã£o em tempo de execuÃ§Ã£o de partes estruturais do programa, lidando apenas com os diferentes dados de entrada em construÃ§Ãµes jÃ¡ compiladas.

## Binary Trees

```python
# binary_trees.py/binary_trees.codon

x_python = [2, 4, 8, 16, 20]
y_python = [0.061, 0.060, 0.108, 1.065, 12.870]

x_pypy = [2, 4, 8, 16, 20]
y_pypy = [0.060, 0.064, 0.085, 1.056, 12.865]

x_codon = [2, 4, 8, 16, 20]
y_codon = [0.0003, 0.0004, 0.0017, 0.744, 12.798]

for i in range(0, len(x_python)):
    plt.plot(x_python[i:], y_python[i:], 'ro-')
    
for i in range(0, len(x_pypy)):
    plt.plot(x_pypy[i:], y_pypy[i:], 'bo-')
    
for i in range(0, len(x_codon)):
    plt.plot(x_codon[i:], y_codon[i:], 'go-')
    
plt.ylabel("Tempo em segundos")
plt.xlabel("Profundidade MÃ¡xima")

python = mpatches.Patch(color='red', label='python')
pypy = mpatches.Patch(color='blue', label='pypy')
codon = mpatches.Patch(color='green', label='codon')

plt.legend(handles=[python, pypy, codon])

plt.figure(figsize=(10, 8))

plt.show()
```

Tratando-se do programa Binary Trees, nota-se que quase nÃ£o hÃ¡ tanta diferenÃ§a entre os dois interpretadores anteriores.

Adicionando o Codon, temos uma pequena melhoria de performance. Isso acontece por ele ser um compilador, transformando cÃ³digo Python em cÃ³digo de mÃ¡quina, e tambÃ©m por ser capaz de realizar multithreading na execuÃ§Ã£o de instruÃ§Ãµes. Para que possa compilar, a tecnologia altera a sintaxe do Python, adicionando tipagem estÃ¡tica.
